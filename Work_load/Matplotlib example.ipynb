```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df = pd.read_excel('work_load.xlsx')
```

```python
s_charge = df['담당자']
```

```python
s_charge_counts = s_charge.value_counts()
```

```python
# 담당자별 소요시간 함계
```

```python
s_charge_counts.plot.pie(autopct='%.2f%%')
plt.title("담당자별 일감 비율")
plt.axis('equal')
plt.show()
```

```python
group = df.groupby('담당자')
```

```python
df4 = group.sum()
```

```python
df4 = df4.sort_values(by='소요시간', ascending = False)
```

```python
df4 = df4[:10]
```

```python
times = df4['소요시간']
```

```python
times.plot(kind='bar', rot=0)
plt.title("담당자별 소요시간 TOP 10")
plt.xlabel("담당자")
plt.ylabel("소요시간")
plt.show()
```
